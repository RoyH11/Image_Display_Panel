# Create a ppt with all the optic disc images for the 1001 patients

```python
# create a new powerpoint presentation
from pptx import Presentation
from pptx.util import Inches
import pydicom
from pydicom.pixel_data_handlers.util import convert_color_space
from PIL import Image
import matplotlib.pyplot as plt
from io import BytesIO
import cv2
import pandas as pd
import os
```

```python
# read in the csv file
path = r"L:\Lab\Roy\AI-READi-Roy\visualization_fundus_optic\initial_ppt\1002_optic.csv"
df = pd.read_csv(path)
df.head()
```

```python
# create a pptx presentation
prs = Presentation()

root_dir = r"L:\Lab\Roy\AI-READi-Roy"


# loop throught the rows, read in the dicom file, and display the image
for index, row in df.iterrows():
    # read in the dicom file
    image_path = os.path.join(root_dir, row['filepath'][1:])
    #print(image_path)
    ds = pydicom.dcmread(image_path)

    # get the image description
    image_description = row['manufacturer'] + "_" + row['manufacturers_model_name'] + "_" + row['laterality'] + "_" + row['anatomic_region'] + "_" + row['imaging']
    print(image_description)

    spacing = ds.PixelSpacing
    image_size_mm = (spacing[0] * ds.Rows, spacing[1] * ds.Columns)

    # get the image color code
    color_code = ds.PhotometricInterpretation

    if color_code == "YBR_FULL_422":
        # convert to RGB
        img = convert_color_space(ds.pixel_array, 'YBR_FULL_422', 'RGB')
    else:
        # convert to RGB
        img = ds.pixel_array

    # save the image as a png in memory byte stream 
    _, buffer = cv2.imencode('.png', cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
    image_stream = BytesIO(buffer)

    # create a new slide
    slide = prs.slides.add_slide(prs.slide_layouts[6])
    # text box for description
    left = Inches(1)
    top = Inches(1)
    width = Inches(8)
    height = Inches(1)
    txBox = slide.shapes.add_textbox(left, top, width, height)
    tf = txBox.text_frame
    tf.text = image_description
    # text box for image size
    left = Inches(1)
    top = Inches(2)
    width = Inches(8)
    height = Inches(1)
    txBox = slide.shapes.add_textbox(left, top, width, height)
    tf = txBox.text_frame
    tf.text = "pixel size: " + str(ds.Rows) + " x " + str(ds.Columns) + " mm size: " + str(image_size_mm[0]) + " x " + str(image_size_mm[1])
    # image
    left = Inches(1)
    top = Inches(3)
    #height_to_width_ratio = ds.Rows / ds.Columns
    pic = slide.shapes.add_picture(image_stream, left, top, width=Inches(4))

    image_stream.close()

# save the presentation
prs.save(r"L:\Lab\Roy\AI-READi-Roy\visualization_fundus_optic\initial_ppt\1002_optic.pptx")
print("done")




    

    

    
```
